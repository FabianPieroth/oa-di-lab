```python
# data import
import scipy.io
import matplotlib.pyplot as plt
```

```python
filename= 'US_data_singleSoS_Scan_12.mat'
```

```python
!ls
```

```python
def plot_us_channel(im_input, im_target, channel):
    plt.figure(figsize=(20,20))
    plt.subplot(1,2,1)
    plt.title('input image')
    plt.imshow(im_input[:,:,channel], cmap = 'gray')


    plt.subplot(1,2,2)
    plt.title('target image')
    plt.imshow(im_target[:,:,channel], cmap = 'gray')
    
```

```python
target_im = scipy.io.loadmat(filename)['US_high']
input_im = scipy.io.loadmat(filename)['US_low']
```

```python
plot_us_channel(input_im, target_im, 0)
```

```python
plot_us_channel(input_im, target_im, 1)
```

```python
plot_us_channel(input_im, target_im, 2)
```

```python
plot_us_channel(input_im, target_im, 3)
```

```python
plot_us_channel(input_im, target_im, 4)
```
